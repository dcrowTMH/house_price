1) high dimensional construction

Data in high dimension and we want to approximate its shape or topology

Each data point is so call 0-simplex

Nerve theorem

connect the 0-simplex and form 1-simplex

We can approximate it into 2-simplex or higher dimension and approximate the topology.

UAMP extend a radius around each point and make a connection between each point and its neighbor with insercting radii

The radii are equal

First problem, Cannot connect the point with larger gap (Low density region)

Too many neighbor (High density region)

Second problem, the curse of dimensionality (distances between point become more and more similar)

-> Using variable radius instead (Riemennian metrics)

UMAP does not estimate the density as a number but uses a proxy:

density was estimated as high if the kth neighbor is close & lower when the kth neighbor is far away.

kth is the hyperparameter that we have to choose. (Help to find the right local radius)

k is big -> more global structure is preserved

k is small -> radius decrease and the local structure is more preserved 

rarely recipe to finding the optimized kth automatically

Some trial & error is needed & kth depends on each data set individually

Different neighbor get different distances & neighbor get a weight and connection probability

Point far away are weighted less & low connection probability

* High dimension graph was constructed and ready to project to low dimension.
